### __Experiment 1__ : Examine the dataset from bank-additional-full.csv.

```python
import pandas as pd

df = pd.read_csv('bank-additional-full.csv', sep = ';')
df.head()
```

The dataset from bank-additional-full.csv is read using the Pandas library. The first few rows of the dataframe are displayed above and show that there are 20 variables used to determine whether a client will subscribe to a term deposit. The final column consists of variable 'y' to indicate whether the individual represented by that row subscribed. This is the value we will be classifying.

### __Experiment 2__ : Split the features and target values into separate training and test sets.

```python
from sklearn.model_selection import train_test_split

training, test = train_test_split(df, train_size = 0.9, test_size = 0.1, random_state=(2021-10-25))

print(f'Size of training set: {len(training)}')
training.head()
```

```python
print(f'Size of test set: {len(test)}')
test.head()
```

The dataset has been separated into a training set and a test set using the scikit-learn library. The training set consists of 90% of the data while the test set is made up of the remaining 10%. A random state is passed in as well to ensure the data is shuffled the same way every time the program runs to ensure reproducibility. To confirm the dataset split properly, the first few rows of both sets as well as the size of each set are displayed above.

### __Experiment 3__ : Process the training and test sets before they are used.

### Experiment 3a : Discard the duration column.

```python
training = training.drop(columns = 'duration')
test = test.drop(columns = 'duration')

print("Columns in training set:")
print(training.columns)

print("\nColumns in test set:")
print(test.columns)
```

The duration of a call drastically impacts the results of the model. For example, if duration = 0, y becomes "no". Because of this, it is discarded to obtain an accurate model. A list of column names for both the training and test sets are shown above to ensure duration does not exist among them.

### Experiment 3b : Set aside feature y for use in training and testing since it is the target response. Then, drop it from the features.

```python
# Storing y column
training_y = training['y']
test_y = test['y']

# Dropping y column from both training and test sets
training = training.drop(columns = 'y')
test = test.drop(columns = 'y')

print("Columns in training set:")
print(training.columns)

print("\nColumns in test set:")
print(test.columns)
```

Column y is the target response we want to find using our model. Because of this, we want to keep it separate from our features. The column names of the training and test sets have been printed once more to show the y column has properly dropped.

### __Experiment 4 :__ Take input variables from bank-additional-names.txt to determine whether a client will subscribe to a term deposit.

```python
# Create subset of training set and test sets using only bank client data
training_inputs = training[['job', 'marital', 'age', 'education', 'default', 'housing', 'loan']]
test_inputs = test[['job', 'marital', 'age', 'education', 'default', 'housing', 'loan']]

# Convert categorical values into dummy variables so that entire dataframe is numeric
training_inputs = pd.get_dummies(training_inputs, columns = ['job', 'marital', 'education', 'default', 'housing', 'loan'], 
                          prefix_sep='.', drop_first = True)
test_inputs = pd.get_dummies(test_inputs, columns = ['job', 'marital', 'education', 'default', 'housing', 'loan'], 
                          prefix_sep='.', drop_first = True)
```

The features that will be used for the model have been identified under "bank client data" in the bank-additional-names.txt file. Out of the seven features, six are categorical. To make them treatable as vectors, we turn them into dummy variables using the Pandas get_dummies() function. We also pass drop_first as True to get k-1 dummies out of the k categorical levels in order to ensure we have no issues with collinear features.

### __Experiment 5 :__ Fit a Categorical Naive Bayes classifier to the features identified in Experiment 4. Then, score it on both the training and test sets to test its accuracy.

```python
import numpy as np
from sklearn.naive_bayes import CategoricalNB
clf = CategoricalNB()

clf.fit(training_inputs, training_y)

print(f"Training Score: {clf.score(training_inputs, training_y)}")
print(f"Test Score: {clf.score(test_inputs, test_y)}")
```

The Categorical Naive Bayes classifier seems to be very accurate. The training and test scores were both approximately .88 indicating a high correlation between the predicted and actual outcomes. The model was fit using the training set and training y values and then tested on both the training and test sets. It is interesting, however, that the values are so close even though the test set had not been introduced to the model and the training set was used to fit the model.

### __Experiment 10__: Create the dumb classifier vector and then create confusion matrix to find AUC.

```python
import numpy as np
from sklearn.metrics import confusion_matrix as cm

# Convert yes and no to 1 and 0 respectively
temp_y = test_y.eq('yes').mul(1)

# Create 0 vector
dummy_y = np.zeros_like(temp_y)

# Create confusion matrix with dummy_y and temp_y
matrix = cm(temp_y.values, dummy_y)

# Find AUC score
from sklearn.metrics import roc_auc_score

print(f"Confusion Matrix:\n {matrix}")
print(f"AUC: {roc_auc_score(temp_y, dummy_y)}")
```

In this experiment we created a dummy vector of zeros equal in length to the test set. We used this to create a confusion matrix and receive the AUC score. The confusion matrix follows the following format:
\begin{bmatrix}
TN & FP \\
FN & TP
\end{bmatrix}
<br>
<center>
    <strong>TP</strong> = True Negative<br>
    <strong>FN</strong> = False Negative<br>
    <strong>FP</strong> = False Positive<br>
    <strong>TP</strong> = True Positive<br>
</center>

This means that our matrix produces 3646 True Negatives. This is accurate since there were 3646 no's found in previous experiment. We also had 473 false negatives. This means that we predicted no (since all the predictions were no) while there were 473 yes's. This is also accurate according to the previous experiment's yes count.

Using the dummy vector and the test outcome values we can compute the Area Under the Curve (AUC) to be 0.5. This is expected since we used a constant prediction value; an AUC of 0.5 indicates that the model really does not predict a value or simply returns a constant. Graphically, this would mean the curve is simple the line y = x indicating a constant slope with half the area of the graph underneath.

### __Experiment 12__: Oversample to add weight to yes values.

```python
# Oversample training set
no_count, yes_count = training_y.value_counts()
training_yes = training_inputs[training_y == 'yes']
training_no = training_inputs[training_y == 'no']
training_yes_y = training_y[training_y == 'yes']
training_no_y = training_y[training_y == 'no']

df_yes = training_yes.sample(no_count, replace=True, random_state=(2021-10-25))
df_yes_y = training_yes_y.sample(no_count, replace=True, random_state=(2021-10-25))

oversampled_training = pd.concat([training_no, df_yes], axis=0)
oversampled_training_y = pd.concat([training_no_y, df_yes_y], axis=0)

# Oversample test set
no_count, yes_count = test_y.value_counts()
test_yes = test_inputs[test_y == 'yes']
test_no = test_inputs[test_y == 'no']
test_yes_y = test_y[test_y == 'yes']
test_no_y = test_y[test_y == 'no']

df_yes = test_yes.sample(no_count, replace=True, random_state=(2021-10-25))
df_yes_y = test_yes_y.sample(no_count, replace=True, random_state=(2021-10-25))

oversampled_test = pd.concat([test_no, df_yes], axis=0)
oversampled_test_y = pd.concat([test_no_y, df_yes_y], axis=0)

oversampled_training[oversampled_training_y == 'yes'].head()
```

In this experiment we oversampled the data by adding more yes values. We know that the yes count is far lower than the no count and we duplicated yes outcomes to grow the training and test sets until there are an equal number of yes and no values. The resulting training set has 32902 no values (unchanged) but now also has 32902 yes values. The mapping is created using the training outcomes vector. In addition, we scaled the test set, training outcomes, and test outcomes so that all four sets have the correct dimensions for further analysis.

### __Experiment 13__: Retrain both classifiers and find the score, confusion matrix, and AUC.

```python
# Categorical Naive Bayes
oversampled_training_age = oversampled_training
oversampled_test_age = oversampled_test
oversampled_training_age['age'] = oversampled_training_age['age'] // 10
oversampled_test_age['age'] = oversampled_test_age['age'] // 10

clf_age = CategoricalNB()
clf_age.fit(oversampled_training_age, oversampled_training_y)

print(f"CategoricalNB - Training Score: {clf_age.score(oversampled_training_age, oversampled_training_y)}")
print(f"CategoricalNB - Test Score: {clf_age.score(oversampled_test_age, oversampled_test_y)}")

temp_y = oversampled_test_y.eq('yes').mul(1)
pred_y = clf_age.predict(oversampled_test_age)
pred_y[pred_y=='yes'] = 1
pred_y[pred_y=='no'] = 0
pred_y = pred_y.astype(int)
matrix = cm(temp_y.values, pred_y)

print(f"CategoricalNB - Confusion Matrix:\n {matrix}")
print(f"CategoricalNB - AUC: {roc_auc_score(temp_y, pred_y)}")

# K-Nearest Neighbors
neigh = knn(n_neighbors = 5)
neigh.fit(oversampled_training, oversampled_training_y)

print(f"\n\nKNN - Training Score: {neigh.score(oversampled_training, oversampled_training_y)}")
print(f"KNN - Test Score: {neigh.score(oversampled_test, oversampled_test_y)}")

pred_y = neigh.predict(oversampled_test)
pred_y[pred_y=='yes'] = 1
pred_y[pred_y=='no'] = 0
pred_y = pred_y.astype(int)

# create confusion matrix for exp 8 with pred_y and temp_y
matrix = cm(temp_y.values, pred_y)
print(f"KNN - Confusion Matrix:\n {matrix}")
print(f"KNN - AUC: {roc_auc_score(temp_y, pred_y)}")
```

After oversampling the data we can see an improvement in both the CategoricalNB classifier as well as the KNN classifier. The AUC score for the CategoricalNB classifier jumped up to .587 and the KNN AUC also jumped up to .579. In addition, the confusion matrix shows many more values for our true positives and true negatives. There are still a significant number of false positives and false negatives, however. Since these values are relatively low even after oversampling the data, the bank client data may not be a good set of features to predict the outcomes. Possible solutions would be to choose different features or select different models (or both).
